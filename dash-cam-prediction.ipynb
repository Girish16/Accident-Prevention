```python
import os
import cv2
import pandas as pd

video_dir = "/kaggle/input/nexar-collision-prediction/train/"

# List all MP4 files in the directory
video_files = [f for f in os.listdir(video_dir) if f.endswith(".mp4")]

# Initialize a list to store metadata
video_data = []

for file in video_files:
    file_path = os.path.join(video_dir, file)
    
    # Open video using OpenCV
    cap = cv2.VideoCapture(file_path)
    if cap.isOpened():
        fps = cap.get(cv2.CAP_PROP_FPS)
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        duration = frame_count / fps if fps > 0 else 0
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        # Store metadata
        video_data.append({
            "file_name": file,
            "file_path": file_path,
            "fps": fps,
            "frame_count": frame_count,
            "duration": duration,
            "resolution": f"{width}x{height}"
        })
    
    cap.release()

# Create a DataFrame
df_videos = pd.DataFrame(video_data)

```

```python
import os
import cv2
import pandas as pd

# Path to dataset directory
video_dir = f'/kaggle/input/nexar-collision-prediction/train'

# Function to extract relevant frames
def extract_relevant_frames(video_id, event_time, save_dir):
    video_id = str(video_id).rjust(5, "0")  # Convert ID to 5-digit format
    video_path = os.path.join(video_dir, f"{video_id}.mp4")

    if not os.path.exists(video_path):
        print(f"Video not found: {video_path}")
        return

    # Open video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error opening video: {video_path}")
        return

    fps = int(cap.get(cv2.CAP_PROP_FPS))  # Get frames per second
    start_frame = max(0, int((event_time - 5) * fps))  # 5 sec before event
    end_frame = int((event_time + 5) * fps)  # 5 sec after event

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break  # End of video

        if start_frame <= frame_count <= end_frame:
            frame_filename = os.path.join(save_dir, f"{video_id}_frame_{frame_count}.jpg")
            cv2.imwrite(frame_filename, frame)

        frame_count += 1

    cap.release()
    print(f"Frames saved for {video_id} from {start_frame} to {end_frame}.")

# Directory to save extracted frames
save_frames_dir = "/kaggle/working/extracted_frames/"
os.makedirs(save_frames_dir, exist_ok=True)

# Loop through dataset and extract frames
for id in ids:

    extract_relevant_frames(id, filtered_data[filtered_data["id"] == int(id)]["time_of_event"].iloc[0], save_frames_dir)
```
